```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ca=pd.read_csv("./linkedin-jobs-canada.csv")
```

```python
# Making new data frame 
all_data = [ca]
df = pd.concat(all_data)
```

```python
df.posted_date = pd.to_datetime(df['posted_date'])
```

```python
company=df.groupby('company')['title'].count()
company.sort_values(ascending=False).head(10)
```

```python
sns.countplot(ca.onsite_remote)
# Add labels and a title
plt.xlabel('Job Types')
plt.ylabel('Posted Jobs')
plt.title('Job Posts vs Job Types in Canada')
# Show the legend
plt.legend()

# Display the bar chart
plt.show()
```

```python
df["onsite_remote"].value_counts().plot(kind="pie",autopct="%.2f")
```

```python
df['title'].unique()
```

```python
plt.figure(figsize=(40,15))
plt.style.use('seaborn-whitegrid')
ax=sns.countplot(x="location",data=df)
plt.xlabel('company')
plt.ylabel('Count')
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.45, p.get_height()+1), va = 'bottom', color = 'black')
     
```

```python
plt.figure(figsize = (15, 7))
company = df["company"].value_counts()[:30]
sns.barplot(x =company, y = company.index, palette = "Accent")
plt.xlabel("Count")
```

```python
plt.figure(figsize = (15, 7))
location = df["location"].value_counts()[:30]
sns.barplot(x =location, y = location.index, palette = "Accent")
plt.xlabel("Count")
```

```python
plt.figure(figsize = (15, 7))
salary = df["salary"].value_counts()[:10]
sns.barplot(x =salary, y = salary.index, palette = "Accent")
plt.xlabel("Count")
```
